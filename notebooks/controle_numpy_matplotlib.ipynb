# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('Iris.csv', delimiter=',', skip_header=1)
print("Shape du dataset:", data.shape)

```

### A2. Séparation des features et des espèces
Séparez :
- les 4 premières colonnes numériques dans un array `features`
- la dernière colonne (espèces) dans un array `species`
Utilisez le slicing.


```python
features = data[:, :4]
species = data[:, 4]

print("Shape features:", features.shape)
print("Shape species:", species.shape)
```

### A3. Indexing
Affichez les 5 premières lignes de `features` en utilisant l’indexing.


```python
print("5 premières lignes de features:")

print(features[:5])
```

### A4. Extraction d’une colonne
Créez un array `sepal_length` contenant uniquement la première colonne de `features`.  
Affichez sa `size` et son `dtype`.


```python
sepal_length = features[:, 0]
print("Size:", sepal_length.size)
print("Dtype:", sepal_length.dtype)
```

### A5. Slicing par espèce
Utilisez le slicing pour extraire les 50 premières lignes correspondant à l’espèce *Setosa*.  
Stockez-les dans `setosa_features`.


```python
setosa_features = features[:50]
print("Shape setosa_features:", setosa_features.shape)

```

# Partie B — Statistiques Basiques et Opérations Mathématiques

### B6. Moyenne par colonne
Calculez la moyenne de chaque colonne de `features` avec `np.mean(axis=0)`.


```python
means = np.mean(features, axis=0)
print("Moyennes par colonne:", means)

```

### B7. Min et Max
Trouvez la valeur minimale et maximale de la longueur des pétales.


```python
petal_length = features[:, 2]
min_petal = np.min(petal_length)
max_petal = np.max(petal_length)

print(f"Min longueur pétales: {min_petal}")
print(f"Max longueur pétales: {max_petal}")
```

Normalisez les longueurs des pétales en les divisant par leur maximum.

```python
normalized_petals = petal_length / max_petal
print("5 premières valeurs normalisées:", normalized_petals[:5])
```

### E23. Grille 2x2
Histogrammes des 4 caractéristiques avec `plt.subplots(2,2)`.


```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
features_names = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

for i, ax in enumerate(axes.flat):
    ax.hist(features[:, i], bins=15, edgecolor='black', alpha=0.7)
    ax.set_xlabel(features_names[i])
    ax.set_ylabel('Fréquence')
    ax.set_title(f'Distribution - {features_names[i]}')
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python
setosa_mean = np.mean(setosa_features, axis=0)
versicolor_mean = np.mean(features[50:100], axis=0)
virginica_mean = np.mean(features[100:], axis=0)

setosa_std = np.std(setosa_features, axis=0)
versicolor_std = np.std(features[50:100], axis=0)
virginica_std = np.std(features[100:], axis=0)

x = np.arange(4)
width = 0.25

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x - width, setosa_mean, width, yerr=setosa_std, label='Setosa', capsize=5)
ax.bar(x, versicolor_mean, width, yerr=versicolor_std, label='Versicolor', capsize=5)
ax.bar(x + width, virginica_mean, width, yerr=virginica_std, label='Virginica', capsize=5)

ax.set_xticks(x)
ax.set_xticklabels(features_names)
ax.set_ylabel('Valeur (cm)')
ax.set_title('Moyennes par Espèce avec Écart-Type')
ax.legend()
ax.grid(axis='y', alpha=0.3)
plt.show()
```

# Partie F — Mini-Analyse Finale

### F27. Corrélations
Calculez la matrice de corrélation avec `np.corrcoef`.


```python
correlation_matrix = np.corrcoef(features.T)
print("Matrice de corrélation:")
print(correlation_matrix)
print("\nFonctionnalités:")
for i, name in enumerate(features_names):
    print(f"{i}: {name}")
```

### F28. Visualisation des corrélations
Visualisez les corrélations sous forme de bar chart ou scatter.


```python
fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)

ax.set_xticks(np.arange(4))
ax.set_yticks(np.arange(4))
ax.set_xticklabels(features_names, rotation=45, ha='right')
ax.set_yticklabels(features_names)

for i in range(4):
    for j in range(4):
        text = ax.text(j, i, f'{correlation_matrix[i, j]:.2f}',
                      ha="center", va="center", color="black")

ax.set_title('Matrice de Corrélation des Features')
plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
setosa_var = np.var(setosa_features)
versicolor_var = np.var(features[50:100])
virginica_var = np.var(features[100:])

print(f"Variance Setosa: {setosa_var:.4f}")
print(f"Variance Versicolor: {versicolor_var:.4f}")
print(f"Variance Virginica: {virginica_var:.4f}")

variances = [setosa_var, versicolor_var, virginica_var]
max_var_idx = np.argmax(variances)
species_names = ['Setosa', 'Versicolor', 'Virginica']
print(f"\nEspèce avec plus grande variance: {species_names[max_var_idx]}")

species_data = [setosa_features, features[50:100], features[100:]][max_var_idx]
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for i, ax in enumerate(axes.flat):
    ax.hist(species_data[:, i], bins=10, edgecolor='black', alpha=0.7)
    ax.set_xlabel(features_names[i])
    ax.set_ylabel('Fréquence')
    ax.set_title(f'{species_names[max_var_idx]} - {features_names[i]}')
    ax.grid(axis='y', alpha=0.3)

plt.suptitle(f'Distributions pour {species_names[max_var_idx]} (Variance maximale)')
plt.tight_layout()
plt.show()
```
